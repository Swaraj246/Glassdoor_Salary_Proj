```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('EDA-Salary Data.csv')
```

```python
#Selecting the relevant models
df_model = df[['average_salary','Rating','Size','Type of ownership','Industry','Sector','Revenue','Comp','hourly','employer_provided',
               'States','Same State','age','Python','Spark','aws','Excel','Tableau','job_simp','seniority','desc_len']]
```

```python
#Creating dummy variables and converting categorical columns into numerical columns
df_dummy = pd.get_dummies(df_model)
```

```python
X = df_dummy.drop('average_salary', axis = 1)
Y = df_dummy.average_salary.values
#Let's split the data into 70% training and 30% testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# Running a multiple linear regression model (OLS) from the statsmodels library
import statsmodels.api as sm
X_sum = X = sm.add_constant(X)
model = sm.OLS(Y,X_sum).fit()
model.summary()
```

```python
# We can also use the scikit-learn library
from sklearn.linear_model import LinearRegression
```

```python
# Linear Regression Model
lm = LinearRegression()
lm.fit(X_train,Y_train)
```

```python
#Cross_validating to select the best training set
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(lm,X_train,Y_train,scoring = 'neg_mean_absolute_error', cv=3))
```

```python
# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,Y_train)
np.mean(cross_val_score(rf,X_train,Y_train,scoring = 'neg_mean_absolute_error', cv=3))
```

```python
# Support vector Machine Regressor
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_train,Y_train)
np.mean(cross_val_score(regressor,X_train,Y_train,scoring = 'neg_mean_absolute_error', cv=3))
```

```python
# Optimizing Random Forest Regressor model using GridsearchCV
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':range(10,300,10), 'criterion':('mse','mae'), 'max_features':('auto','sqrt','log2')}
gs = GridSearchCV(rf,parameters,scoring='neg_mean_absolute_error',cv=3)
gs.fit(X_train, Y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

```python
# testing ensembles models
model_pred_lm = lm.predict(X_test)
model_pred_rf = rf.predict(X_test)
model_pred_svr = regressor.predict(X_test)
```

```python
# Comparing it with Mean Absolute Error
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(Y_test,model_pred_rf)
```

```python
mean_absolute_error(Y_test,model_pred_svr)
```

```python
mean_absolute_error(Y_test,model_pred_lm)
```
